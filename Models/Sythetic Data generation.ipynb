```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(4)

# Number of data points per clsas
num_samples = 200

# Generate synthetic data for Class 1
mean_class1 = [2, 2]  # Mean of Class 1
cov_class1 = [[1, 0.5], [0.5, 1]]  # Covariance matrix of Class 1
class1_data = np.random.multivariate_normal(mean_class1, cov_class1, num_samples)

# Generate synthetic data for Class 2
mean_class2 = [-2, -2]  # Mean of Class 2
cov_class2 = [[1, -0.5], [-0.5, 1]]  # Covariance matrix of Class 2
class2_data = np.random.multivariate_normal(mean_class2, cov_class2, num_samples)

# Create labels for the classes (0 for Class 1, 1 for Class 2)
class1_labels = np.zeros(num_samples)
class2_labels = np.ones(num_samples)

# Combine the data and labels for the entire dataset
synthetic_data = np.vstack((class1_data, class2_data))
synthetic_labels = np.hstack((class1_labels, class2_labels))

# Visualize the synthetic data
plt.scatter(synthetic_data[:, 0], synthetic_data[:, 1], c=synthetic_labels)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Data')
plt.show()
```
